### 문제 답지 분리
- y = ax + b
- X(문제): loc, iloc
- y(답지): loc, iloc

- iloc[행시작:행끝, 열시작:열끝]




```python
X= data.iloc[:,1:] # 문제
y= data.iloc[:,0] # 답지
```

```python
print('원본 ',data.shape, '문제(X) ',X.shape,'답지(y) ',y.shape)
```

### 원 핫 인코딩
- 문자를 숫자로 변형시켜주는 함수
 - 범주형 -> 수치형으로 변환
- 문자로 학습할수 없기 때문에 숫자로 변환 시킴

```python
# get_dummies(X) : X에 들어있는 모든 특징들을 원핫인코딩해주세요
# astype(int) : 원핫인코딩된 값들을 숫자형태(정수)로 바꿔주세요
X_ohe = pd.get_dummies(X).astype(int)
X_ohe
```

```python
print('X-원본',X.shape,'X-원핫' ,X_ohe.shape)
```
