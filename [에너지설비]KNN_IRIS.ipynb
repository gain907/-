### 2. 데이터 수집

```python
# sklearn에서 제공하는 데이터셋 활용
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python
iris_data.keys()
```

```python
# data:특징데이터(문제데이터)
# target:특징데이터(답지데이터)
# target_names : 답지데이터 이름
# feature_names : 문제데이터 이름
```

```python
# 문제데이터(특징)
iris_data.data
```

```python
# 문제데이터(답지)
iris_data.target
```

```python
# 문제데이터(답지 이름)
iris_data.target_names
```

```python
# 문제데이터(문제데이터 이름)
iris_data.feature_names
```

```python
import pandas as pd

# 문제 데이터 만들기
X = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

```python
X.info()
```

```python
X.head() # 기본 5개 데이터 뽑아주세요
```

### 4. EDA(탐색적 데이터 분석)
 - 산점도

```python
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기
pd.plotting.scatter_matrix(X, #그래프에 표현될 데이터
                           figsize=(15,15), # 그래프 크기
                           c = iris_data.target, # 답지종류에 따른 색 적용
                           marker = 'v',# 점의 모양
                           alpha=  0.7 # 투명도
                           )
plt.show() # 그래프로 보여주세요
```
