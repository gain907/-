### 반복문
- for 변수 in 반복가능한 객체:

```python
# range = 인덱스 카운트
for i in range(3):
   print(i)
```

```python
for i in [1,3,5,7,9]:
  print(i)

```

```python
# 리스트를 변수에 담아서 반복문 돌리기
리스트변수 = [1,3,5,7,9]
for i in 리스트변수:
  print(i)
```

```python
리스트변수 = [1,"3","5","7","9"]
for i in 리스트변수:
  print(i)
```

```python
# for문에 조건문 만들기
# 만약 1이면 '1' 출력
# 만약 9이면 '9' 출력
# 나머지는 '출력없음' 출력
# 결과값은 1,출력없음,출력없음,출력없음,9
# 같으면 == ex) i == 1

리스트변수 = [1,3,5,7,9]
for i in 리스트변수:
  if i == 1:
    print("1")
  elif i == 9:
    print("9")
  else:
    print("출력없음")
```

```python
# for문으로 구구단?
for i in [2,3,4,5,6,7,8,9]:
  for j in [1,2,3,4,5,6,7,8,9]:
    # print(i,j)
    print("{}x{}={}".format(i,j,i * j))
```
