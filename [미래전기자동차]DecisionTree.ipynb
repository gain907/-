### 문제(X),답지(y) 분리
- X(문제): loc,iloc 이용해서
- y(답지): loc,iloc 이용해서


```python
# 문제데이터 : class를 제외한 모든 컬럼
# 답지데이터 : class 컬럼
X = data.loc[:,'cap-shape':]
# x = data.iloc[:,1:]

y = data.loc[:,'class']
#y = data.iloc[:,0]
```

### 원핫 인코딩
- pd.get_dummies
- 문자로 학습했을때 학습이 잘되지 않음
- 값을 수치화 시켜주는 것이 필요
- 범주형->수치형

```python
# X : 원본데이터
X_ohe = pd.get_dummies(X)
X_ohe = X_ohe.astype(int)  # 숫자타입으로 바꿔주세요
```
