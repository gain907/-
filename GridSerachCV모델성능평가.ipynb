```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/폴리텍전남캠퍼스/data/mushrooms.csv')
```

```python
# 데이터 정보확인
data.shape
```

```python
X = data.loc[:,'cap-shape':]
# x = data.iloc[:,1:]

y = data.loc[:,'class']
#y = data.iloc[:,0]
```

```python
data.shape,X.shape,y.shape
```

```python
# X : 원본데이터
X_ohe = pd.get_dummies(X)
X_ohe = X_ohe.astype(int)  # 숫자타입으로 바꿔주세요
```

```python
X_ohe
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_ohe,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape, X_test.shape)
```

### 모델 성능평가 GrdiSerachCV
- 교차검증과 최적의 파라미터 튜닝을 동시 진행

```python
# 모델 생성
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
# 모델 객체 생성
cvtree_model = DecisionTreeClassifier()
```

```python
# max_depth   3,5,7
# max_leaf_nodes 4,6,8
# min_samples_split 1000,2000,3000
# min_samples_leaf 400,500,600
params = {'max_depth':[3,5,7],
          'max_leaf_nodes':[4,5,6],
          'min_samples_split':[1000,2000,3000],
          'min_samples_leaf':[400,500,600]
          }
```

```python
grid_tree = GridSearchCV(cvtree_model, param_grid=params, cv=5, refit=True)
```

```python
# param_grid의 설정한 파라미터를 순차적으로 학습/평가
grid_tree.fit(X_train,y_train)
```

```python
grid_tree.cv_results_
```

```python
grid_result = pd.DataFrame(grid_tree.cv_results_)
grid_result
```

```python
grid_result.info()
```

```python
# best_params_
print('최적의 하이퍼 파라미터',grid_tree.best_params_)
print('최고 교차 검증 점수', grid_tree.best_score_)
print('최고 성능 모델', grid_tree.best_estimator_)
```

```python
best_model = DecisionTreeClassifier(max_depth=3, max_leaf_nodes=4, min_samples_leaf=400,
                       min_samples_split=1000)
```

```python
best_model.fit(X_train,y_train)
```

```python
best_pred = best_model.predict(X_test)
best_pred
```

```python
# accuracy_score
from sklearn.metrics import accuracy_score
import numpy as np
best_score = accuracy_score(y_test,best_pred)
print('최적모델의 정확도 ',np.round(best_score*100,2))
```
